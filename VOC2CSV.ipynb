# get data from xml files and write data into a csv file
# for xml files in all_path

```python
#coding=utf-8
import xml.dom.minidom
import csv
import numpy as np
import os
```

```python
all_path = os.walk('./验证/')  #for your own dir****
for (root, dirs, files) in all_path:
    for j in files:
        print(j)
        #open the xml file and get data
        dom = xml.dom.minidom.parse('./验证/'+j) #for your own dir ****
        root_csv = dom.documentElement
        filenameTag = root_csv.getElementsByTagName('filename') #filename
        filename1 = filenameTag[0]
        filename = filename1.firstChild.data
        bb = root_csv.getElementsByTagName('object')
        for elm in bb:
            xminTag = elm.getElementsByTagName('xmin') #xmin
            xmin1 = xminTag[0]
            xmin = xmin1.firstChild.data
            xmaxTag = elm.getElementsByTagName('xmax') #xmax
            xmax1 = xmaxTag[0]
            xmax = xmax1.firstChild.data
            yminTag = elm.getElementsByTagName('ymin') #ymin
            ymin1 = yminTag[0]
            ymin = ymin1.firstChild.data
            ymaxTag = elm.getElementsByTagName('ymax')#ymax
            ymax1 = ymaxTag[0]
            ymax = ymax1.firstChild.data
            name = elm.getElementsByTagName('name')#ymax
            name = name[0]
            name = name.firstChild.data
            print (filename, xmin, xmax, ymin, ymax, name)  
            cols=1
            datacsv = np.zeros((cols,5)).astype(np.int32)
            data_file_name = np.zeros((cols,1)).astype(np.object)#the name of jpg 
            for i in range(cols):
                data_file_name[i,0] = filename 
                datacsv[i,4] = name
                datacsv[i,0] = xmin
                datacsv[i,1] = xmax
                datacsv[i,2] = ymin
                datacsv[i,3] = ymax
            data_concat = np.concatenate((data_file_name,datacsv),axis=1)
            #print(data_concat)
            with open("./验证/xml_cv2.csv","a") as csvfile:   #for your own dir and csv file name***
                writer = csv.writer(csvfile)
                writer.writerows(data_concat)

                

 
```

```python
with open("./验证/xml_cv2.csv",'rt')as fin:  #读有空行的csv文件，舍弃空行  
    #for your own dir and csv file name  ***
    lines=''  
    for line in fin:  
        if line!='\n':  
            lines+=line        

with open("./验证/xml_cv2.csv",'wt')as fout:  #再次文本方式写入，不含空行  
     #for your own dir and csv file name  ***
    fout.write(lines)
```
